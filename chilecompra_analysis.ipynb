# Análisis del Mercado de Compras Públicas en Chile (2019-2025)

**Autor:** Guillermo Fuentes Jaque (Científico de Datos Geoespaciales)
**Fecha:** 2026-03-04

## 1. Introducción

Este informe presenta un análisis exhaustivo del sistema de compras públicas de Chile, utilizando datos abiertos de **Mercado Público (ChileCompra)**. El objetivo es identificar patrones, tendencias y oportunidades de mercado a través de la exploración de datos de licitaciones y órdenes de compra desde 2019 hasta principios de 2025.

El análisis abarca:
- **Evolución del gasto público:** Análisis de la serie temporal del gasto total y número de licitaciones.
- **Análisis sectorial y regional:** Desglose del gasto por sector económico y región geográfica.
- **Identificación de oportunidades:** Creación de un índice de oportunidad de mercado para identificar las categorías más atractivas.
- **Análisis de competencia:** Evaluación del número de oferentes por licitación en diferentes categorías.
- **Modelamiento predictivo:** Pronóstico de la demanda futura de gasto público utilizando modelos de series temporales.

## 2. Metodología

El proyecto se desarrolló siguiendo un pipeline de ciencia de datos estándar:

1.  **Recolección de Datos:** Se descargaron datos masivos de licitaciones (2019-2025) desde el portal de datos abiertos de ChileCompra. Se procesaron más de 15 millones de registros.
2.  **Procesamiento y ETL:** Los datos crudos (más de 30 GB en formato CSV) se procesaron en streaming para extraer información relevante, limpiar y normalizar los datos, y generar datasets agregados en formato Parquet, optimizando el uso de memoria y disco.
3.  **Ingeniería de Variables:** Se calcularon métricas clave como el crecimiento anual del gasto, la competencia promedio y un índice de oportunidad de mercado.
4.  **Análisis Exploratorio y Visualización:** Se generaron visualizaciones para explorar las tendencias y patrones en los datos.
5.  **Modelamiento Predictivo:** Se implementaron modelos Prophet y XGBoost para pronosticar el gasto futuro.

## 3. Análisis Exploratorio de Datos

### 3.1. Evolución del Gasto Total en Compras Públicas

El gasto en compras públicas muestra una fuerte estacionalidad, con picos de gasto hacia fin de año. Se observa un peak anómalo en 2020, probablemente asociado a gastos de emergencia por la pandemia de COVID-19. A partir de 2021, el número de licitaciones mensuales se ha mantenido relativamente estable, aunque con fluctuaciones.

![Evolución del Gasto Total](figures/fig01_evolucion_gasto_total.png)

### 3.2. Gasto por Sector Económico

El sector de **Salud** es consistentemente el de mayor gasto, seguido por **Gobierno Central** y **Municipalidades**. Esto refleja las prioridades del gasto público en Chile.

![Gasto por Sector](figures/fig02_gasto_anual_sector.png)

### 3.3. Top Categorías de Licitación

Las categorías con mayor monto adjudicado son **Servicios de Construcción y Mantenimiento** y **Obras**, lo que indica una inversión constante en infraestructura. **Obras MINVU** también representa una porción significativa del gasto.

![Top Categorías](figures/fig03_top_categorias.png)

### 3.4. Distribución Geográfica del Gasto

La **Región Metropolitana** concentra la mayor parte del gasto público, lo cual es esperable dada su densidad poblacional y actividad económica. Le siguen a distancia las regiones de **Coquimbo** y **Atacama**.

![Gasto por Región](figures/fig04_gasto_region.png)

### 3.5. Nivel de Competencia

El nivel de competencia (número promedio de oferentes por licitación) varía significativamente entre categorías. Algunas áreas, como **Servicios de Publicidad y Marketing**, muestran una alta competencia, mientras que otras, como **Equipamiento y Suministros Médicos**, tienen menos oferentes, lo que podría indicar barreras de entrada o mercados más especializados.

![Nivel de Competencia](figures/fig05_competencia.png)

## 4. Identificación de Oportunidades de Mercado

Para identificar las áreas más prometedoras, se construyó un **Índice de Oportunidad de Mercado**. Este índice combina tres factores:
1.  **Alto Gasto:** Categorías con un alto volumen de gasto total.
2.  **Alto Crecimiento:** Categorías con un crecimiento porcentual significativo en el último año.
3.  **Baja Competencia:** Categorías con un número reducido de oferentes.

El análisis revela que **Obras** y **Vehículos y Equipamiento en General** presentan una oportunidad de mercado **Alta** o **Muy Alta**, debido a su combinación de alto gasto y crecimiento reciente.

![Índice de Oportunidad](figures/fig06_indice_oportunidad.png)

## 5. Modelamiento Predictivo

Se utilizó el modelo **Prophet** de Facebook para pronosticar la serie temporal del gasto total en compras públicas para los próximos 3 años (2025-2028). El modelo captura la tendencia general y la estacionalidad anual de los datos.

La predicción sugiere que, si bien se mantendrá la estacionalidad, no se esperan los picos de gasto observados durante la pandemia. El gasto se estabilizaría en un nivel base con fluctuaciones predecibles a lo largo del año.

![Forecast del Gasto Total](figures/fig07_forecast_total.png)

## 6. Trazabilidad del Gasto: Flujo de Inversión Pública (Sankey)

Para comprender la dinámica del gasto público de manera integral, se generó un diagrama Sankey que visualiza el flujo de inversión desde los organismos compradores, pasando por las categorías de licitación y los proveedores, hasta las regiones donde se ejecuta el gasto. Este diagrama permite identificar de un vistazo las principales rutas del dinero público y las interconexiones entre los actores clave.

![Análisis de Flujo Masivo de Inversión Pública 2024 (Top 15 por Nivel)](figures/fig13_sankey_flujo_mercado_masivo.png)

## 7. Conclusiones

- El mercado de compras públicas en Chile es un ecosistema dinámico con patrones estacionales claros y una fuerte concentración del gasto en sectores y regiones específicas.
- El análisis del **Índice de Oportunidad de Mercado** permite a los proveedores identificar estratégicamente las categorías con mayor potencial de negocio.
- Los modelos predictivos como Prophet son herramientas valiosas para anticipar la demanda futura y planificar recursos.
- Este proyecto demuestra el poder de los datos abiertos para generar insights accionables y fomentar la transparencia y eficiencia en el gasto público.
